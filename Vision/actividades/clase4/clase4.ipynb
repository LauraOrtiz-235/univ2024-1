```python
import skimage
import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.transform import rotate, rescale, resize
```

```python
img = data.coffee()
plt.imshow(img)
plt.show()

img_1 = rotate(img, angle=45)
plt.imshow(img_1)
plt.show()
```

```python
img_2 = rescale(img, 1/2, anti_aliasing=True, multichannel=True)
plt.imshow(img_2)
plt.show()

img_3 = rescale(img, 5, anti_aliasing=True, multichannel=True)
plt.imshow(img_3)
plt.show()
```

```python
img_4 = resize(img, (200, 300), anti_aliasing=True, anti_aliasing_sigma=1)
plt.imshow(img_4)
plt.show()
```

```python
img2 = plt.imread('rotada.jpg')
plt.imshow(img2)
plt.show()

img2_1 = rotate(img2, angle=-90)
plt.imshow(img2_1)
plt.show()
```
