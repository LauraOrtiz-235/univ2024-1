```python
import skimage
import numpy as np
import matplotlib.pyplot as plt

from skimage import color
from skimage.morphology import binary_dilation, binary_erosion, erosion, dilation
from skimage.util import random_noise
from skimage.filters import gaussian
```

```python
img1 = plt.imread('binaria.jpg')
plt.imshow(img1, cmap='gray')
plt.show()

img1_1 = binary_erosion(img1)
plt.imshow(img1_1, cmap='gray')
plt.show()
```

```python
img1_2 = erosion(img1)
plt.imshow(img1_2, cmap='gray')
plt.show()

img1_3 = dilation(img1)
plt.imshow(img1_3, cmap='gray')
plt.show()
```

```python
img2 = plt.imread('binaria2.jpeg')
img2_gray = color.rgb2gray(img2)
plt.imshow(img2_gray, cmap='gray')
plt.show()

img2_1 = binary_erosion(img2_gray)
plt.imshow(img2_1, cmap='gray')
plt.show()
```

```python
img3 = plt.imread('binaria3.jpg')
img3_gray = color.rgb2gray(img3)
plt.imshow(img3_gray, cmap='gray')
plt.show()

img3_1 = binary_dilation(img3_gray)
plt.imshow(img3_1, cmap='gray')
plt.show()
```

```python
img3_2 = dilation(img3_gray)
plt.imshow(img3_2, cmap='gray')
plt.show()

img3_3 = erosion(img3_gray)
plt.imshow(img3_3, cmap='gray')
plt.show()
```
