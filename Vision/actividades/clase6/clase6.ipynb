```python
from skimage import color, data, util
from skimage.restoration import denoise_bilateral, denoise_wavelet, denoise_nl_means, denoise_tv_bregman, denoise_tv_chambolle 
import matplotlib.pyplot as plt
from skimage.segmentation import slic, chan_vese, expand_labels, felzenszwalb, quickshift
from skimage.color import label2rgb
from skimage.feature import canny
from skimage.filters import sobel, roberts, scharr
```

```python
img1 = data.coffee()
plt.imshow(img1)
plt.show()
```

```python
img1_1 = denoise_bilateral(img1, multichannel=True, sigma_color=1)
plt.imshow(img1_1)
plt.show()
```

```python
img1_2 = denoise_wavelet(img1, multichannel=True, sigma=1)
plt.imshow(img1_2)
plt.show()
```

```python
img1_3 = denoise_nl_means(img1, patch_size=10, patch_distance=10, multichannel=True)
plt.imshow(img1_3)
plt.show()
```

```python
img1_4 = denoise_tv_bregman(img1, weight=0.8, multichannel=True)
plt.imshow(img1_4)
plt.show()

img1_5 = denoise_tv_chambolle(img1, weight=0.5, multichannel=True)
plt.imshow(img1_5)
plt.show()
```

```python
img1_6 = slic(img1, n_segments=100, start_label=1)

img1_7 = label2rgb(img1_6, img1, kind='avg', bg_label=0)
plt.imshow(img1_7)
plt.show()
```

```python
img3 = plt.imread('fotografo.jpg')
plt.imshow(img3, cmap='gray')
plt.show()

img3_gray = color.rgb2gray(img3)
```

```python
img3_1 = slic(img3, n_segments=1000, start_label=1)
img3_2 = label2rgb(img3_1, img3, bg_label=0)
plt.imshow(img3_2)
plt.show()

img3_3 = chan_vese(img3_gray, mu=0.50, lambda1=0.8, lambda2=0.8, max_iter=4000)
plt.imshow(img3_3, cmap='gray')
plt.show()
```

```python
img4 = plt.imread('pergamino2.jpg')
img4_gray = color.rgb2gray(img4)

img4_1 = chan_vese(img4_gray, mu=0.50, lambda1=0.8, lambda2=0.8, max_iter=400)
plt.imshow(img4_1, cmap='gray')
plt.show()
```

```python
img4_2 = canny(img4_gray)
plt.imshow(img4_2, cmap='gray')
plt.show()
```

```python
img4_3 = sobel(img4_gray)
plt.imshow(img4_3, cmap='gray')
plt.show()
```

```python
img5 = data.coins()
img5_gray = color.rgb2gray(img5)

img5_1 = canny(img5_gray)
plt.imshow(img5_1, cmap='gray')
plt.show()

img5_2 = sobel(img5_gray)
plt.imshow(img5_2, cmap='gray')
plt.show()

img5_3 = roberts(img5_gray)
plt.imshow(img5_3, cmap='gray')
plt.show()

img5_4 = scharr(img5_gray)
plt.imshow(img5_4, cmap='gray')
plt.show()
```

```python
img5_slic = slic(img5, n_segments=100, start_label=1)
img5_5 = label2rgb(img5_slic, img5, bg_label=0)
#img5_5 = img5_5 / img5_5.max()

img5_6 = expand_labels(img5_5, distance=300)
plt.imshow(img5_6, cmap='gray')
plt.show()
```

```python
img6 = data.astronaut()
img6_gray = color.rgb2gray(img6)

img6_1 = slic(img6, n_segments=30, start_label=1)
img6_2 = label2rgb(img6_1, img6, kind='avg', bg_label=0)
img6_2 = img6_2 / img6_2.max()
plt.imshow(img6_2)
plt.show()
```

```python
img6_frez = label2rgb(felzenszwalb(img6, scale=100, min_size=35), img6, kind='avg', bg_label=0)
img6_frez = img6_frez / img6_frez.max()
plt.imshow(img6_frez)
plt.show()
```
