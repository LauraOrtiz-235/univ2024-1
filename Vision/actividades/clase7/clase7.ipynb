### Watershed segmentation

```python
img15 = data.astronaut()
img15_gray = rgb2gray(img15)

img15_roberts = roberts(img15_gray)
img15_scharr = scharr(img15_gray)
img15_sobel = sobel(img15_gray)

plt.figure(figsize=(12, 8))
plt.subplot(1, 3, 1)
plt.imshow(img15_roberts, cmap='gray')
plt.title('Roberts')
plt.subplot(1, 3, 2)
plt.imshow(img15_scharr, cmap='gray')
plt.title('Scharr')
plt.subplot(1, 3, 3)
plt.imshow(img15_sobel, cmap='gray')
plt.title('Sobel')
```

```python
img15_ws = watershed(img15_sobel, markers=10)
plt.imshow(img15_ws, cmap='gray')
plt.show()
```
