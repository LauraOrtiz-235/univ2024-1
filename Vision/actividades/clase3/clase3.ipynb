```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage import data
from skimage.filters import try_all_threshold, threshold_otsu, threshold_local, threshold_minimum, threshold_mean, roberts, sobel, scharr, prewitt, gaussian, median
from skimage import color, exposure
```

```python
img = plt.imread('texto.jpg')
img = color.rgb2gray(img)

ax, fig = try_all_threshold(img, figsize=(14, 14), verbose=False)
```

```python
img2 = plt.imread('texto2.jpg')
img2_gray = color.rgb2gray(img2)

ax, fig = try_all_threshold(img2_gray, figsize=(14, 14), verbose=False)
```

```python
img2_1 = exposure.equalize_hist(img2_gray)
plt.imshow(img2_1, cmap='gray')
plt.show()
```

```python
img3 = data.astronaut()
plt.imshow(img3)
plt.show()

img3_gray = color.rgb2gray(img3)
plt.imshow(img3_gray, cmap='gray')
plt.show()
```

```python
img3_1 = exposure.equalize_hist(img3_gray)
plt.imshow(img3_1, cmap='gray')
plt.show()
```

```python
img3_2 = img3_gray > threshold_otsu(img3_gray)
plt.imshow(img3_2, cmap='gray')
plt.show()
```

```python
img3_3 = img3_gray > threshold_local(img3_gray, block_size=15, offset=0)
plt.imshow(img3_3, cmap='gray')
plt.show()

img3_4 = img3_gray > threshold_minimum(img3_gray)
plt.imshow(img3_4, cmap='gray')
plt.show()

img3_5 = img3_gray > threshold_mean(img3_gray)
plt.imshow(img3_5, cmap='gray')
plt.show()
```

```python
img3_6 = roberts(img3_gray)
plt.imshow(img3_6, cmap='gray')
plt.show()

img3_7 = sobel(img3_gray)
plt.imshow(img3_7, cmap='gray')
plt.show()

img3_8 = scharr(img3_gray)
plt.imshow(img3_8, cmap='gray')
plt.show()

img3_9 = prewitt(img3_gray)
plt.imshow(img3_9, cmap='gray')
plt.show()
```

```python
img3_10 = gaussian(img3, sigma=0)
plt.imshow(img3_10)
plt.show()

img3_11 = median(img3_gray)
plt.imshow(img3_11, cmap='gray')
plt.show()
```

```python
img4 = plt.imread('radiografia.jpg')
img4_gray = color.rgb2gray(img4)

img4_1 = exposure.equalize_hist(img4_gray)
plt.imshow(img4_1, cmap='gray')
plt.show()

plt.hist(img4_1.ravel(), bins=256)
plt.show()
```

```python
img4_2 = exposure.equalize_adapthist(img4_gray, clip_limit=0.01)
plt.imshow(img4_2, cmap='gray')
plt.show()

plt.hist(img4_2.ravel(), bins=256)
plt.show()
```

```python
img5 = plt.imread('cerebro.jpeg')
img5_gray = color.rgb2gray(img5)

img5_1 = exposure.equalize_hist(img5_gray)
img5_2 = exposure.equalize_adapthist(img5_gray, clip_limit=0.01)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(img5_1, cmap='gray')
ax[0].title.set_text('Equalize hist')
ax[1].imshow(img5_2, cmap='gray')
ax[1].title.set_text('Equalize adapthist')
plt.show()
```
