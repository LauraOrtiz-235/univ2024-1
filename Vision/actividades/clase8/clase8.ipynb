```python
import matplotlib.pyplot as plt
from skimage import data
from skimage.color import label2rgb, rgb2gray
from skimage import color
from skimage.feature import canny, corner_moravec, corner_harris, corner_peaks
```

```python
img1 = plt.imread('patrones.jpg')
img1_gray = rgb2gray(img1)

img1_canny = canny(img1_gray, sigma=0.05)

plt.figure(figsize=(12, 12))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(img1_canny, cmap='gray')
plt.title('Canny')
plt.show()
```

### Corner Moravec y Harris

```python
img4 = plt.imread('building_image.jpg')
img4_gray = rgb2gray(img4)

img4_moravec = corner_moravec(img4_gray, window_size=2)
img4_canny = canny(img4_gray, sigma=1)
img4_harris = corner_harris(img4_gray, method='eps', eps=0.0001)

plt.figure(figsize=(16, 12))
plt.subplot(1, 4, 1)
plt.imshow(img4, cmap='gray')
plt.title('Original')
plt.subplot(1, 4, 2)
plt.imshow(img4_moravec, cmap='gray')
plt.title('Moravec')
plt.subplot(1, 4, 3)
plt.imshow(img4_canny, cmap='gray')
plt.title('Canny')
plt.subplot(1, 4, 4)
plt.imshow(img4_harris, cmap='gray')
plt.title('Harris')
plt.show()
```

```python
img4_peaks = corner_peaks(img4_harris, min_distance=140)

def show_image_corners(image, corners):
    plt.figure(figsize=(10, 8))
    plt.imshow(image, cmap='gray', interpolation='nearest')
    plt.plot(corners[:, 1], corners[:, 0], 'r+')
    plt.show()

show_image_corners(img4_harris, img4_peaks)
```
