```python
from skimage.feature import Cascade
import matplotlib.pyplot as plt
from skimage import data
```

```python
entrenar = data.lbp_frontal_face_cascade_filename()
```

```python
detector = Cascade(entrenar)
```

```python
img1 = data.astronaut()
```

```python
detectar = detector.detect_multi_scale(img1, scale_factor=1.2, step_ratio=1, min_size=(30, 30), max_size=(200, 200))
```

```python
plt.imshow(img1)
img_mod = plt.gca()
plt.gray()

for patch in detectar:
    img_mod.add_patch(plt.Rectangle((patch['c'], patch['r']), patch['width'], patch['height'], fill=False, edgecolor='r', linewidth=2))
```
