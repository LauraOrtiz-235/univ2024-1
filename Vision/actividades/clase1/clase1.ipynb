```python
import skimage
import matplotlib.pyplot as plt
import numpy as np
from skimage import color
from skimage.filters import try_all_threshold
```

```python
img1 = plt.imread('imagen1.jpeg')
plt.imshow(img1)
plt.show()
```

```python
img1_v1 = np.fliplr(img1)
plt.imshow(img1_v1)
```

```python
img1_v2 = np.flipud(img1)
plt.imshow(img1_v2)
```

```python
img1_gray = color.rgb2gray(img1)
plt.imshow(img1_gray, cmap='gray')
plt.show()

print(img1_gray.shape)
```
