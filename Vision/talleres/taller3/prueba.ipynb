```python
import os
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.segmentation import flood_fill
import numpy as np
```

```python
carpeta_imagenes = '/home/laura/Desktop/univ/univ2024-1/Vision/talleres/taller3/Taller3'

# Obtener la lista de archivos en la carpeta
archivos = [archivo for archivo in os.listdir(carpeta_imagenes) if archivo.endswith(('.jpg', '.jpeg', '.JPG'))]

# Iterar sobre cada archivo de imagen
for archivo in archivos:
    # Ruta completa de la imagen
    ruta_imagen = os.path.join(carpeta_imagenes, archivo)

    # Cargar la imagen
    imagen = io.imread(ruta_imagen)

    # Convertir la imagen a escala de grises si es necesario
    if len(imagen.shape) == 3:
        imagen_gris = color.rgb2gray(imagen)
    else:
        imagen_gris = imagen

    # Aplicar el filtro flood_fill
    resultado = flood_fill(imagen_gris, (0, 0), new_value=1)

    # Mostrar la imagen original y la imagen con el filtro aplicado
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(imagen, cmap='gray')
    plt.title('Imagen Original')

    plt.subplot(1, 2, 2)
    plt.imshow(resultado, cmap='gray')
    plt.title('Resultado con flood_fill')

    plt.show()
```
